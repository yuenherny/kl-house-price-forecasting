# Data Exploration and Cleaning: Transactions

```python
import pandas as pd
from helpers import RAW_DATA_DIR
```

## Exploration

```python
df_landed = pd.read_excel(RAW_DATA_DIR / 'transactions_landed_KL.xlsx')
df_landed
```

```python
df_nonlanded_1 = pd.read_excel(RAW_DATA_DIR / 'transactions_non-landed_KL_1.xlsx')
df_nonlanded_1
```

```python
df_nonlanded_2 = pd.read_excel(RAW_DATA_DIR / 'transactions_non-landed_KL_2.xlsx')
df_nonlanded_2
```

```python
print(df_landed.columns)
print(df_nonlanded_1.columns)
print(df_nonlanded_2.columns)
```

Features in both landed and non-landed datasets:
1. project_name
2. spa_date
3. address
4. building_type
5. tenure
6. rooms
7. price_psf
8. price
9. built_up or lot_size

Features only in landed dataset:
1. floors
2. land_area

Transformation steps:
1. Rename lot_size to built_up
2. For non-landed, floor is set at 1, as high rise properties usually only has one floor
3. For non-landed, land_area is equal to built_up or lot_size as there are no more extra space for except the unit itself

```python
df_nonlanded = pd.concat([df_nonlanded_1, df_nonlanded_2], axis=0)
df_nonlanded
```

```python
df_nonlanded = df_nonlanded.rename(columns={'lot_size': 'built_up'})
df_nonlanded.columns
```

```python
df_nonlanded['floors'] = 1
df_nonlanded.columns
```

```python
df_nonlanded['land_area'] = df_nonlanded['built_up']
df_nonlanded.head()
```

```python
df_transactions = pd.concat([df_landed, df_nonlanded], axis=0)
df_transactions
```

```python
df_transactions['floors'] = df_transactions['floors'].astype(str)
df_transactions['rooms'] = df_transactions['rooms'].astype(str)
df_transactions['land_area'] = df_transactions['land_area'].astype(str)
df_transactions['built_up'] = df_transactions['built_up'].astype(str)
df_transactions['price'] = df_transactions['price'].astype(str)
```

```python
df_transactions.to_parquet(RAW_DATA_DIR / 'transactions_KL.parquet', index=False)
```

```python
df_transactions.to_excel(RAW_DATA_DIR / 'transactions_KL.xlsx', index=False)
```

We use the saved Parquet file to perform data cleaning and transformation.
